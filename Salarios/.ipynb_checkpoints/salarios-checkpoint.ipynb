# Análisis de salarios

#### Variables
- salario = salario nominal
- sexo = 0 hombre, 1 mujer
- edad = edad del encuestado
- años de educ = años de educación.
- nivel de educ = 0 Ninguno 1 Primaria, 2 Secundaria, 3 Superior, 4 Otros.
- experiencia = Años de experiencia
- horas semana = horas trabajadas en la semana
- cant pers empresa = cantidad personal en la empresa donde uno trabaja
- año = año de la encuesta
- ipc = indice de precios al consumidor anual acumulado
- sueldoMin = Sueldo mínimo.
- salarioReal = Salario real (base = 2016)  

#### Objetivo:
Predecir el salario de un hombre con 20 años de educacion y 2 años de experiencia.

En este análisis utilizaremos datos de extraidos de las encuesta del INE, en el periodo 2013 al 2021.

Para ello se realizará lo siguente:

1. Rransformación de datos.
2. Análisis descriptivo. 
3. Predicción del salario.
4. Evaluación de los modelos.
5. Elección del mejor modelo.

```python
# LIBRERIAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
# DATA
salario = pd.read_csv('./data/salario.csv')
```

```python
print(salario.info())
```

```python
print(salario.head())
```

## Transformación de datos

```python
salario['sexo'].unique()
```

```python
# Convertir la columna sexo a datos categoricos
salario['sexo'] = salario['sexo'].astype('category')

salario['sexo'] = salario['sexo'].replace({'Hombre': 0, 'Mujer': 1})
```

```python
# Observar datos únicos de la columna nivel de educ
print(salario['nivel de educ'].unique())

# Borrar filas de la columna "nivel de educ" que contengan la palabra Otros. 
# Dado que se convertira estos datos a categoricos ordinales.
salario = salario[salario['nivel de educ'] != 'Otros']

# Cambiar el tipo de columna "salario" a categórico
salario['nivel de educ'] = salario['nivel de educ'].astype('category')

# Ordenar correctamente estos datos categóricos
salario['nivel de educ'] = salario['nivel de educ'].cat.reorder_categories(['Ninguno','Primaria','Secundaria','Superior'])

# Codificar categorías con valores enteros
salario['nivel de educ'] = salario['nivel de educ'].cat.codes
```

```python
# Cambiar de float a int la columna "exp tiempo"
salario['exp tiempo'] = salario['exp tiempo'].astype(int)

# Cambiar el nombre de columa "exp tiempo" a "experiencia"
salario = salario.rename(columns={'exp tiempo': 'experiencia'})
```

```python
# Cambiuar de float a int la columna "horas semana"
salario['horas semana'] = salario['horas semana'].astype(int)
```

```python
# Cambiar columna año a formato datetime
# salario['año'] = pd.to_datetime(salario['año'], format='%Y')
```

## Análisis descriptivo

### Análisis de datos Atípicos

```python
# Describir de los datos.
print(salario.describe())
```

Comparando la media y la mediana no se ve diferencias grandes en la mayoria de las variables. Tratemos de analizarlas con boxplots.

```python
# Dibujar Boxplots

# Lista de columnas numéricas para las que quieres hacer boxplots
columnas_numericas = ['salario', 'edad', 'años de educ', 'experiencia', 'horas semana', 'cant pers empresa', 'ipc', 'sueldoMin', 'salarioReal']

# Dividir la lista de columnas en dos
mitad = len(columnas_numericas) // 2
columnas_numericas1 = columnas_numericas[:mitad]
columnas_numericas2 = columnas_numericas[mitad:]

# Asegúrarse de que ambas listas tengan la misma longitud
if len(columnas_numericas1) < len(columnas_numericas2):
    columnas_numericas1.append(None)

# Dibujar los boxplots en dos columnas
fig, axs = plt.subplots(len(columnas_numericas1), 2, figsize=(10, 10))

for i, columna in enumerate(columnas_numericas1):
    if columna is not None:
        sns.boxplot(x=salario[columna], ax=axs[i, 0])
        axs[i, 0].set_title(columna)

for i, columna in enumerate(columnas_numericas2):
    if columna is not None:
        sns.boxplot(x=salario[columna], ax=axs[i, 1])
        axs[i, 1].set_title(columna)

plt.tight_layout()
plt.show()
```

- Segun la norma de jornada laboral de trabajo, una persona sólo puede trabajar 48 a la semana. Dado que muchas personas tiene dos trabajos, tendremos un rango superior de 48*2 = 96 horas a la semana. Por lo que borraremos de la columna "horas semana". los que exedan de 96 horas.

```python
# Crear una condición booleana donde 'horas semana' es menor o igual a 96
condicion = salario['horas semana'] <= 96

# Usar la condición para filtrar el DataFrame
salario = salario[condicion]
```

- Con respecto a la columna "experiencia" borrar el dato 500

```python
# Crear una condición booleana donde 'experiencia' es diferente de 500
condicion = salario['experiencia'] != 500

# Usar la condición para filtrar el DataFrame
salario = salario[condicion]
```

### Análisis de componentes principales

```python
X = salario[['sexo','sueldoMin', 'experiencia', 'horas semana', 'años de educ']]
y = salario['salarioReal']

# Crear el objeto PCA
pca = PCA(n_components=4)

# Ajustar y transformar los datos de entrada
X_pca = pca.fit_transform(X)

# Los componentes principales se encuentran en pca.components_
print("Componentes principales: ", pca.components_)

# La varianza explicada por cada componente principal se encuentra en pca.explained_variance_ratio_
print("Varianza explicada: ", pca.explained_variance_ratio_)
```

Claro, aquí está una interpretación más detallada de los resultados del Análisis de Componentes Principales (PCA):

1. **Componente Principal 1 (CP1)**: Este componente está dominado por la variable `sueldoMin`. Un cambio en `sueldoMin` resultará en un gran cambio en CP1. Este componente explica aproximadamente el **99.56%** de la varianza en los datos.

2. **Componente Principal 2 (CP2)**: Este componente está dominado por la variable `horas semana`. Un cambio en `horas semana` resultará en un cambio notable en CP2. Este componente explica aproximadamente el **0.29%** de la varianza en los datos.

3. **Componente Principal 3 (CP3)**: Este componente está dominado por la variable `experiencia`. Un cambio en `experiencia` resultará en un cambio notable en CP3. Este componente explica aproximadamente el **0.13%** de la varianza en los datos.

4. **Componente Principal 4 (CP4)**: Este componente está dominado por la variable `años de educ`. Un cambio en `años de educ` resultará en un cambio notable en CP4. Este componente explica aproximadamente el **0.02%** de la varianza en los datos.

En resumen, la mayoría de la varianza en los datos puede ser explicada por `sueldoMin`, seguido por `horas semana`, `experiencia` y `años de educ`. Estos resultados pueden ser útiles para entender las relaciones entre estas variables y para reducir la dimensionalidad de los datos para análisis posteriores. Sin embargo, ten en cuenta que los componentes principales son menos interpretables que las variables originales.
