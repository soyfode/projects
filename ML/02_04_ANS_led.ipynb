# Explorar el conjunto de datos de dígitos LED

Utilizaremos NMF para descomponer imágenes en escala de grises en sus patrones más comunes. 

En primer lugar, exploraremos el conjunto de datos de imágenes y veremos cómo se codifica como una matriz. 

Se tiene 100 imágenes como muestras de un array 2D, donde cada fila representa una única imagen de 13x8. Las imágenes de su conjunto de datos son imágenes de una pantalla digital LED.

```python
# LIBRERIAS
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF
```

```python
# DATA
ruta_archivo = '/home/jovyan/notebooks/data/lcd-digits.csv'
samples = pd.read_csv(ruta_archivo,index_col = None, header = None)
samples = samples.values
samples.shape
```

```python
# Seleccione la fila 0: dígit
digit = samples[0,:]

# Imprimir digit
print(digit)

# Usar el método .reshape() de digit para obtener un array 2D con forma (13, 8). 
# Asigna el resultado a bitmap.
bitmap = digit.reshape((13,8))

# Imprinit bitmap
print(bitmap)

# Dibujar bitmap
plt.imshow(bitmap, cmap = "gray", interpolation='nearest')
plt.colorbar()
plt.show()
```

# El NMF aprende las partes de las imágenes

Ahora descompongamos el conjunto de datos de dígitos. 

```python
# FUNCIONES
## Muestra la imagen codificada por cualquier matriz 1D:
def show_as_image(sample):
    bitmap = sample.reshape((13, 8))
    plt.figure()
    plt.imshow(bitmap, cmap='gray', interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
# Crear un modelo NMF: model
model = NMF(n_components = 7)

# Aplicar fit_transform para las muestras: features
features = model.fit_transform(samples)

# Llamar show_as_image en cada componente
for component in model.components_:
    show_as_image(component)

# Seleccionar la fila 0 de funciones: digit_features
digit_features = features[0,:]

# Imprimir digit_features
print(digit_features)
```

## PCA no aprende las partes

A diferencia del NMF, el PCA no aprende las partes de las cosas. Sus componentes no corresponden a temas (en el caso de los documentos) o a partes de imágenes, cuando se entrena en imágenes.
