```python
# LIBRERIAS
from IPython.display import Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from math import floor, ceil
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error as MSE
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
# FUNCIONES
def plot_decision_regions(X, y, clf,
                          feature_index=None,
                          filler_feature_values=None,
                          filler_feature_ranges=None,
                          ax=None,
                          X_highlight=None,
                          zoom_factor=1.,
                          legend=1,
                          hide_spines=True,
                          markers='s^oxv<>',
                          colors=('blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan'),
                          scatter_kwargs=None,
                          contourf_kwargs=None,
                          scatter_highlight_kwargs=None):
    dim = X.shape[1]

    if ax is None:
        ax = plt.gca()

    plot_testdata = True
    if not isinstance(X_highlight, np.ndarray):
        if X_highlight is not None:
            raise ValueError('X_highlight must be a NumPy array or None')
        else:
            plot_testdata = False
    elif len(X_highlight.shape) < 2:
        raise ValueError('X_highlight must be a 2D array')

    if feature_index is not None:
        # Unpack and validate the feature_index values
        if dim == 1:
            raise ValueError(
                'feature_index requires more than one training feature')
        try:
            x_index, y_index = feature_index
        except ValueError:
            raise ValueError(
                'Unable to unpack feature_index. Make sure feature_index '
                'only has two dimensions.')
        try:
            X[:, x_index], X[:, y_index]
        except IndexError:
            raise IndexError(
                'feature_index values out of range. X.shape is {}, but '
                'feature_index is {}'.format(X.shape, feature_index))
    else:
        feature_index = (0, 1)
        x_index, y_index = feature_index

    # Extra input validation for higher number of training features
    if dim > 2:
        if filler_feature_values is None:
            raise ValueError('Filler values must be provided when '
                             'X has more than 2 training features.')

        if filler_feature_ranges is not None:
            if not set(filler_feature_values) == set(filler_feature_ranges):
                raise ValueError(
                    'filler_feature_values and filler_feature_ranges must '
                    'have the same keys')

        # Check that all columns in X are accounted for
        column_check = np.zeros(dim, dtype=bool)
        for idx in filler_feature_values:
            column_check[idx] = True
        for idx in feature_index:
            column_check[idx] = True
        if not all(column_check):
            missing_cols = np.argwhere(~column_check).flatten()
            raise ValueError(
                'Column(s) {} need to be accounted for in either '
                'feature_index or filler_feature_values'.format(missing_cols))

    marker_gen = cycle(list(markers))

    n_classes = np.unique(y).shape[0]
    colors_gen = cycle(colors)
    colors = [next(colors_gen) for _ in range(n_classes)]

    # Get minimum and maximum
    x_min, x_max = (X[:, x_index].min() - 1./zoom_factor,
                    X[:, x_index].max() + 1./zoom_factor)
    if dim == 1:
        y_min, y_max = -1, 1
    else:
        y_min, y_max = (X[:, y_index].min() - 1./zoom_factor,
                        X[:, y_index].max() + 1./zoom_factor)

    xnum, ynum = plt.gcf().dpi * plt.gcf().get_size_inches()
    xnum, ynum = floor(xnum), ceil(ynum)
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, num=xnum),
                         np.linspace(y_min, y_max, num=ynum))

    if dim == 1:
        X_predict = np.array([xx.ravel()]).T
    else:
        X_grid = np.array([xx.ravel(), yy.ravel()]).T
        X_predict = np.zeros((X_grid.shape[0], dim))
        X_predict[:, x_index] = X_grid[:, 0]
        X_predict[:, y_index] = X_grid[:, 1]
        if dim > 2:
            for feature_idx in filler_feature_values:
                X_predict[:, feature_idx] = filler_feature_values[feature_idx]
    Z = clf.predict(X_predict.astype(X.dtype))
    Z = Z.reshape(xx.shape)
    # Plot decision region
    # Make sure contourf_kwargs has backwards compatible defaults
    contourf_kwargs_default = {'alpha': 0.45, 'antialiased': True}
    if contourf_kwargs is None:
        contourf_kwargs = contourf_kwargs_default
    else:
        contourf_kwargs_default.update(contourf_kwargs)
        contourf_kwargs = contourf_kwargs_default
    cset = ax.contourf(xx, yy, Z,
                       colors=colors,
                       levels=np.arange(Z.max() + 2) - 0.5,
                       **contourf_kwargs)

    ax.contour(xx, yy, Z, cset.levels,
               colors='k',
               linewidths=0.5,
               antialiased=True)

    ax.axis([xx.min(), xx.max(), yy.min(), yy.max()])

    # Scatter training data samples
    # Make sure scatter_kwargs has backwards compatible defaults
    scatter_kwargs_default = {'alpha': 0.8, 'edgecolor': 'black'}
    if scatter_kwargs is None:
        scatter_kwargs = scatter_kwargs_default
    else:
        scatter_kwargs_default.update(scatter_kwargs)
        scatter_kwargs = scatter_kwargs_default
    for idx, c in enumerate(np.unique(y)):
        if dim == 1:
            y_data = [0 for _ in X[y == c]]
            x_data = X[y == c]
        elif dim == 2:
            y_data = X[y == c, y_index]
            x_data = X[y == c, x_index]
        elif dim > 2 and filler_feature_ranges is not None:
            class_mask = y == c
            feature_range_mask = get_feature_range_mask(
                X, filler_feature_values=filler_feature_values,
                filler_feature_ranges=filler_feature_ranges)
            y_data = X[class_mask & feature_range_mask, y_index]
            x_data = X[class_mask & feature_range_mask, x_index]
        else:
            continue

        ax.scatter(x=x_data,
                   y=y_data,
                   c=colors[idx],
                   marker=next(marker_gen),
                   label=c,
                   **scatter_kwargs)

    if hide_spines:
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    if dim == 1:
        ax.axes.get_yaxis().set_ticks([])

    if plot_testdata:
        if dim == 1:
            x_data = X_highlight
            y_data = [0 for _ in X_highlight]
        elif dim == 2:
            x_data = X_highlight[:, x_index]
            y_data = X_highlight[:, y_index]
        else:
            feature_range_mask = get_feature_range_mask(
                X_highlight, filler_feature_values=filler_feature_values,
                filler_feature_ranges=filler_feature_ranges)
            y_data = X_highlight[feature_range_mask, y_index]
            x_data = X_highlight[feature_range_mask, x_index]

        # Make sure scatter_highlight_kwargs backwards compatible defaults
        scatter_highlight_defaults = {'c': 'none',
                                      'edgecolor': 'black',
                                      'alpha': 1.0,
                                      'linewidths': 1,
                                      'marker': 'o',
                                      's': 80}
        if scatter_highlight_kwargs is None:
            scatter_highlight_kwargs = scatter_highlight_defaults
        else:
            scatter_highlight_defaults.update(scatter_highlight_kwargs)
            scatter_highlight_kwargs = scatter_highlight_defaults
        ax.scatter(x_data,
                   y_data,
                   **scatter_highlight_kwargs)

    if legend:
        if dim > 2 and filler_feature_ranges is None:
            pass
        else:
            handles, labels = ax.get_legend_handles_labels()
            ax.legend(handles, labels,
                      framealpha=0.3, scatterpoints=1, loc=legend)

    return ax


def plot_labeled_decision_regions(X, y, models):
    
    # Verificar si la lista de modelos contiene solo dos clasificadores
    if len(models) != 2:
        raise Exception('Los modelos deben ser una lista que contenga solo dos clasificadores entrenados.')
    
    # Verificar si X es un DataFrame de pandas con dos caracterÃ­sticas numÃ©ricas
    if not isinstance(X, pd.DataFrame):
        raise Exception('X debe ser un DataFrame de pandas con dos caracterÃ­sticas numÃ©ricas.')
    
    # Verificar si y es una Series de pandas correspondiente a las etiquetas
    if not isinstance(y, pd.Series):
        raise Exception('y debe ser una Series de pandas correspondiente a las etiquetas.')
    
    # Crear subtramas
    fig, ax = plt.subplots(1, 2, figsize=(6.0, 2.7), sharey=True)
    
    # Iterar sobre los modelos y trazar las regiones de decisiÃ³n
    for i, model in enumerate(models):
        plot_decision_regions(X.values, y.values, model, legend=2, ax=ax[i])
        ax[i].set_title(model.__class__.__name__)
        ax[i].set_xlabel(X.columns[0])
        
        if i == 0:
            ax[i].set_ylabel(X.columns[1])
        
        ax[i].set_ylim(X.values[:, 1].min(), X.values[:, 1].max())
        ax[i].set_xlim(X.values[:, 0].min(), X.values[:, 0].max())
    
    # Ajustar el diseÃ±o y mostrar el grÃ¡fico
    plt.tight_layout()
    plt.show()
```

## ClasificaciÃ³n y Ã¡rboles de regresiÃ³n

### Ãrbol de decisiones para la clasificaciÃ³n

#### Ãrbol de clasificaciÃ³n
Dado un conjunto de datos etiquetado, **un Ã¡rbol de clasificaciÃ³n aprende una secuencia de preguntas sobre caracterÃ­sticas individuales para inferir las etiquetas**. 
A diferencia de los modelos lineales, 
- los Ã¡rboles pueden **capturar relaciones no lineales entre caracterÃ­sticas y etiquetas**.
- AdemÃ¡s, los Ã¡rboles **no requieren que las caracterÃ­sticas estÃ©n en la misma escala mediante la estandarizaciÃ³n**

#### Conjunto de datos sobre cÃ¡ncer de mama en 2D
Para comprender los Ã¡rboles de manera mÃ¡s concreta, intentaremos predecir si un tumor es maligno o benigno en el conjunto de datos de cÃ¡ncer de mama de Wisconsin utilizando solo dos caracterÃ­sticas. La figura aquÃ­ muestra un diagrama de dispersiÃ³n de dos caracterÃ­sticas de cÃ©lulas cancerosas con tumores malignos en azul y tumores benignos en rojo.

```python
# Crear un mapa de colores basado en el diagnÃ³stico
colores = {'M':'red', 'B':'blue'} # Cambia 'M' y 'B' a los valores Ãºnicos en tu columna de diagnÃ³stico

plt.figure(figsize=(10, 6))
plt.scatter(wbc['radius_mean'], wbc['concave points_mean'], c = wbc['diagnosis'].map(colores))

plt.title('Diagrama de dispersiÃ³n de radius_mean vs concave points_mean')
plt.xlabel('radius_mean')
plt.ylabel('concave points_mean')
plt.show()
```

#### Diagrama de Ã¡rbol de decisiones
**Cuando se entrena un Ã¡rbol de clasificaciÃ³n en este conjunto de datos, el Ã¡rbol aprende una secuencia de preguntas if-else y cada pregunta involucra una caracterÃ­stica y un punto de divisiÃ³n**. Habitualmente en el grÃ¡fico de un diagrma de Ã¡rbol, en la parte superior, el Ã¡rbol pregunta si la media de puntos cÃ³ncavos de una instancia es $\leq 0.051$. Si es asÃ­, la instancia atraviesa la rama Verdadera; de lo contrario, atraviesa la rama Falso. De manera similar, la instancia sigue atravesando las ramas internas hasta llegar al final. Luego se predice que la etiqueta de la instancia serÃ¡ la de la clase predominante en ese extremo. ***El nÃºmero mÃ¡ximo de ramas que separan la parte superior de un extremo se conoce como profundidad mÃ¡xima.**

#### Ãrbol de clasificaciÃ³n en scikit-learn
Ahora que sabes quÃ© es un Ã¡rbol de clasificaciÃ³n, ajustemos uno con scikit-learn. 

Primero, importe DecisionTreeClassifier desde sklearn.tree como se muestra en la lÃ­nea 1. AdemÃ¡s, importe las funciones `train_test_split()` desde `sklearn.model_selection` y `precision_score()` desde `sklearn.metrics`. Para obtener una estimaciÃ³n imparcial del rendimiento de un modelo, debe evaluarlo en un conjunto de pruebas invisible. 

Para hacerlo, primero divida los datos en 80% de entrenamiento y 20% de prueba usando train_test_split(). Establezca el parÃ¡metro estratificar en y para que los conjuntos de entrenamiento y de prueba tengan la misma proporciÃ³n de etiquetas de clase que el conjunto de datos no dividido. Ahora puede utilizar `DecisionTreeClassifier()` para crear una instancia de un clasificador de Ã¡rbol, dt con una profundidad mÃ¡xima de 2 estableciendo el parÃ¡metro `max_ Depth` en 2. Tenga en cuenta que el parÃ¡metro `random_state` se establece en 1 para mayor reproducibilidad.

Luego llame al mÃ©todo de ajuste en `dt` y pase `X_train` e `y_train`. Para predecir las etiquetas del conjunto de prueba, llame al mÃ©todo de predicciÃ³n en dt. Finalmente imprima la precisiÃ³n del conjunto de prueba usando `precision_score()`. Para comprender las predicciones del Ã¡rbol de manera mÃ¡s concreta, veamos cÃ³mo clasifica las instancias en el espacio de caracterÃ­sticas.

```python
# Dividir variables dependientes de independiente
X = wbc[['radius_mean','concave points_mean']]
y = wbc['diagnosis']

# Dividir datos en 80% de entrenamiento y 20% de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.2,
                                                   stratify = y,
                                                   random_state = 1)

# Instanciar dt
dt = DecisionTreeClassifier(max_depth = 2, random_state = 1)

# Ajustar el modelo a los datos
dt.fit(X_train, y_train)

# Predecir con los datos de prueba
y_pred = dt.predict(X_test)

# Evaluar la precisiÃ³n del conjunto de datos de prueba
acc = accuracy_score(y_test, y_pred)

print("PrecisiÃ³n del conjunto de prueba: {:.2f}".format(acc))

# Imprimir las primeras 5 predicciones
print(y_pred[0:5])
```

### Aprendizaje del Ã¡rbol de clasificaciÃ³n
Veremos cÃ³mo un Ã¡rbol de clasificaciÃ³n aprende de los datos.

#### Componentes bÃ¡sicos de un Ã¡rbol de decisiones
Primero comencemos definiendo algunos tÃ©rminos. Un Ã¡rbol de decisiÃ³n es una estructura de datos que ***consta de una jerarquÃ­a de unidades individuales llamadas nodos. Un nodo es un punto que implica una pregunta o una predicciÃ³n.***

#### Componentes bÃ¡sicos de un Ã¡rbol de decisiones
- ***La raÃ­z*** es el nodo en el que el Ã¡rbol de decisiÃ³n comienza a crecer. No tiene nodo padre e implica una pregunta que da lugar a 2 nodos hijos a travÃ©s de dos ramas.
- ***Un nodo interno*** es un nodo que tiene un padre. TambiÃ©n implica una pregunta que da lugar a 2 nodos hijos.
- ***Un nodo que no tiene hijos se llama hoja***. Una hoja tiene un nodo padre y no implica preguntas. Es donde se hace una predicciÃ³n.

Recuerde que cuando un Ã¡rbol de clasificaciÃ³n se entrena en un conjunto de datos etiquetado, el Ã¡rbol aprende patrones de las caracterÃ­sticas de tal manera que produzca las hojas mÃ¡s puras. En otras palabras, el Ã¡rbol estÃ¡ formado de tal manera que, en cada hoja, predomine una etiqueta de clase.

#### PredicciÃ³n
Imaginemonos un diagrama de Ã¡rbol, considere el caso en el que una instancia atraviesa el Ã¡rbol para llegar a la hoja de la izquierda. En esta hoja, hay 257 casos clasificados como benignos y 7 casos clasificados como malignos. Como resultado, la predicciÃ³n del Ã¡rbol para este caso serÃ­a: "benigna". Para comprender cÃ³mo un Ã¡rbol de clasificaciÃ³n produce las hojas mÃ¡s puras posibles, primero definamos el concepto de ganancia de informaciÃ³n.

#### Ganancia de informaciÃ³n (IG)
Los nodos de un Ã¡rbol de clasificaciÃ³n crecen de forma recursiva; es decir, ***la obtenciÃ³n de un nodo interno o de una hoja depende del estado de sus antecesores***. Para producir las hojas mÃ¡s puras posibles, en cada nodo, un Ã¡rbol hace una pregunta que involucra una caracterÃ­stica f y un punto de divisiÃ³n sp. Pero, *Â¿cÃ³mo sabe quÃ© caracterÃ­stica y quÃ© punto de divisiÃ³n elegir?* ***Lo hace maximizando la ganancia de informaciÃ³n*** El Ã¡rbol considera que cada nodo contiene informaciÃ³n y tiene como objetivo maximizar la ganancia de informaciÃ³n obtenida despuÃ©s de cada divisiÃ³n. Considere el caso en el que un nodo con N muestras se divide en un nodo izquierdo con N muestras izquierdas y un nodo derecho con N muestras derechas. La ganancia de informaciÃ³n para dicha divisiÃ³n viene dada por la fÃ³rmula que se muestra aquÃ­. 
$$IG(f,sp)=I(parent)-\left(\dfrac{N_{left}}{N}I(left)+\dfrac{N_{right}}{N}\right)i(right)$$
Una pregunta que quizÃ¡s tengas en mente aquÃ­ es: *Â¿QuÃ© criterio se utiliza para medir la impureza de un nodo?* Bueno, existen diferentes criterios que puedes utilizar entre los que se encuentran:
- el Ã­ndice de Gini
- la entropia
Ahora que sabemos quÃ© es la ganancia de informaciÃ³n, describamos cÃ³mo aprende un Ã¡rbol de clasificaciÃ³n.

#### Aprendizaje del Ã¡rbol de clasificaciÃ³n
Cuando se entrena un Ã¡rbol sin restricciones, los nodos crecen de forma recursiva. En otras palabras, un nodo existe segÃºn el estado de sus predecesores. En un nodo que no es hoja, los datos se dividen segÃºn la caracterÃ­stica `f` y el punto de divisiÃ³n `sp` de tal manera que se maximice la ganancia de informaciÃ³n. Si la ganancia de informaciÃ³n obtenida al dividir un nodo es nula, el nodo se declara hoja. Tenga en cuenta que estas reglas son para Ã¡rboles sin restricciones. Si limita la profundidad mÃ¡xima de un Ã¡rbol a 2, por ejemplo, todos los nodos que tengan una profundidad de 2 se declararÃ¡n hojas incluso si la ganancia de informaciÃ³n obtenida al dividir dichos nodos no es nula.

#### Criterio de informaciÃ³n en scikit-learn (conjunto de datos sobre cÃ¡ncer de mama)
Al revisar el conjunto de datos de cÃ¡ncer de mama 2D de la lecciÃ³n anterior, puede establecer el criterio de informaciÃ³n de `dt` en el Ã­ndice de gini estableciendo el parÃ¡metro de criterio en 'gini' como se muestra en la Ãºltima lÃ­nea aquÃ­.

Ahora ajuste dt al conjunto de entrenamiento y prediga las etiquetas del conjunto de prueba. Luego determine la precisiÃ³n del conjunto de prueba de dt, que se evalÃºa en aproximadamente 92%.

```python
# Instanciar dt, para gini.
dt = DecisionTreeClassifier(criterion = 'gini', random_state = 1)

# Ajustar dt al conjunto de entrenamiento
dt.fit(X_train, y_train)

# Predecir el conjunto de prueba
y_pred = dt.predict(X_test)

# Evaluar la precesiÃ³n del conjunto de prueba
accuracy_score(y_test, y_pred)
```

#### Usando la entropÃ­a como criterio
Ahora, entrenaremos un Ã¡rbol de clasificaciÃ³n en el conjunto de datos de cÃ¡ncer de mama de Wisconsin utilizando la entropÃ­a como criterio de informaciÃ³n. Lo harÃ¡ utilizando las 30 funciones del conjunto de datos, que se divide en un 80 % de entrenamiento y un 20 % de prueba.

```python
# Dividir variables dependientes de independiente
X = wbc.drop('diagnosis', axis = 1).values
y = wbc['diagnosis'].values

# Dividir datos en 80% de entrenamiento y 20% de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.2,
                                                   stratify = y,
                                                   random_state = 1)

# Instanciar dt_entropy, establecer 'entropy' como el criterio de informaciÃ³n
dt_entropy = DecisionTreeClassifier(max_depth = 8, criterion = 'entropy', random_state=1)

# Ajustar dt_entropy al conjunto de entrenamiento
dt_entropy.fit(X_train, y_train)

# Instancier dt_gini , establecer 'gini' como el criterio de informaciÃ³n
dt_gini = DecisionTreeClassifier(max_depth = 8, criterion = 'gini', random_state=1)

# Ajustar dt_entropy al conjunto de entrenamiento
dt_gini.fit(X_train, y_train)
```

#### EntropÃ­a vs Ã­ndice de Gini
Compararemos la precisiÃ³n del conjunto de pruebas `dt_entropy` con la precisiÃ³n de  `dt_gini`. 

**NOTA** El indice de Gini es mÃ¡s rÃ¡pido de calcular.

```python
# Utilizar dt_entropy para predecir las etiquetas del conjunto de prueba
y_pred = dt_entropy.predict(X_test)

# Evaluar accuracy_entropy
accuracy_entropy = accuracy_score(y_test, y_pred)

# Utilizar dt_gini para predecir las etiquetas del conjunto de prueba
y_pred1 = dt_gini.predict(X_test)

# Evaluar accuracy_entropy
accuracy_gini = accuracy_score(y_test, y_pred1)

# Imprimir accuracy_entropy
print(f'PrecisiÃ³n lograda al usar entropÃ­a: {accuracy_entropy:.3f}')

# Imprimir accuracy_gini
print(f'PrecisiÃ³n lograda al usar el Ã­ndice de gini: {accuracy_gini:.3f}')
```

### Ãrbol de decisiones para la regresiÃ³n
Veremos cÃ³mo entrenar un Ã¡rbol de decisiÃ³n para un problema de regresiÃ³n. Recuerde que en la regresiÃ³n la variable objetivo es continua. En otras palabras, el resultado de su modelo es un valor real.

#### Conjunto de datos de mpg automÃ¡tico
Motivemos nuestra discusiÃ³n sobre la regresiÃ³n presentando el conjunto de datos de millas por galÃ³n de automÃ³viles del Repositorio de aprendizaje automÃ¡tico de UCI. Este conjunto de datos consta de 6 caracterÃ­sticas correspondientes a las caracterÃ­sticas de un automÃ³vil y una variable objetivo continua denominada mpg que significa millas por galÃ³n. *Nuestra tarea es predecir el consumo de mpg de un automÃ³vil dadas estas seis caracterÃ­sticas*. Para simplificar el problema, aquÃ­ el anÃ¡lisis se restringe a una sola caracterÃ­stica correspondiente al desplazamiento de un automÃ³vil. Esta caracterÃ­stica se indica mediante displ.

```python
# Variable dependientes y independiente
X = auto['displ'].values
y = auto['mpg'].values
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

#### Auto-mpg con una funciÃ³n
Un diagrama de dispersiÃ³n 2D de mpg versus displ muestra que el consumo de mpg disminuye de forma no lineal con el desplazamiento. Tenga en cuenta que los modelos lineales como la regresiÃ³n lineal no podrÃ­an capturar una tendencia no lineal. Veamos cÃ³mo se puede entrenar un Ã¡rbol de decisiones con scikit-learn para resolver este problema de regresiÃ³n.

```python
# Dibujar un diagrma de dispersiÃ³n
plt.scatter(X,y)
plt.xlabel("displ")
plt.ylabel("mpg")
plt.show()
```

#### Ãrbol de regresiÃ³n en scikit-learn
Primero, importe DecisionTreeRegressor de `sklearn.tree` y las funciones `train_test_split()` de `sklearn.model_selection` y `mean_squared_error` como `MSE()` de `sklearn.metrics`. Luego, divida los datos en 80% de tren y 20% de prueba usando `train_test_split`. Ahora puede crear una instancia de `DecisionTreeRegressor()` con una profundidad mÃ¡xima de 4 estableciendo el parÃ¡metro max_profundidad en 4. AdemÃ¡s, establezca el parÃ¡metro `min_sample_leaf` en 0.1 para imponer una condiciÃ³n de detenciÃ³n en la que cada hoja debe contener al menos 10 % de los datos de entrenamiento.

Ahora ajuste `dt` al conjunto de entrenamiento y prediga las etiquetas del conjunto de prueba. Para obtener el error cuadrÃ¡tico medio de su modelo en el conjunto de prueba; proceder de la siguiente manera: 
- primero evaluar el error cuadrÃ¡tico medio.
- luego elevar el valor obtenido a la potencia 1/2.
- imprima el conjunto de prueba `rmse` de `dt` para obtener un valor de 5.1

```python
# Dividir el conjunto en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.2,
                                                   random_state = 3)

# Instanciar DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth = 4,
                          min_samples_leaf = 0.1,
                          random_state = 3)

# Ajustar el modelo a los datos de entrenamiento
dt.fit(X_train, y_train)

# Predecir con el conjunto de datos de prueba
y_pred = dt.predict(X_test)

# Calcular MSE del conjunto de prueba
mse_dt = MSE(y_test, y_pred)

# Calcular la raiz cuadra de mse_dt
rmse_dt = mse_dt**(1/2)

# Imprimir rms_dt
print(rmse_dt)
```

#### Criterio de informaciÃ³n para el Ã¡rbol de regresiÃ³n
$$I(nodo) = MSE(nodo) = \dfrac{1}{N_{nodo}}\sum_{i\in nodo}\left(y^{(i)}-\hat{y}_{nodo}\right)^2$$
$$\hat{y}_{nodo} = \dfrac{1}{N_{nodo}}\sum_{i\in nodo}y^{(i)}$$
AquÃ­, es importante tener en cuenta que, cuando se entrena un Ã¡rbol de regresiÃ³n en un conjunto de datos, la impureza de un nodo se mide utilizando el error cuadrÃ¡tico medio de los objetivos en ese nodo. Esto significa que **el Ã¡rbol de regresiÃ³n intenta encontrar las divisiones que producen hojas donde en cada hoja los valores objetivo son en promedio, lo mÃ¡s cercano posible al valor medio de las etiquetas en esa hoja en particular.**

#### PredicciÃ³n
Cuando una nueva instancia atraviesa el Ã¡rbol y llega a una determinada hoja, su variable objetivo 'y' se calcula como el promedio de las variables objetivo contenidas en esa hoja, como se muestra en esta fÃ³rmula.
$$\hat{y}_{pred}(hoja) = \dfrac{1}{N_{hoja}}\sum_{i\in hoja} y^{(i)}.$$

#### RegresiÃ³n lineal versus Ã¡rbol de regresiÃ³n
Para resaltar la importancia de la flexibilidad de los Ã¡rboles de regresiÃ³n, observe esta figura. A la izquierda tenemos un diagrama de dispersiÃ³n de los datos en azul junto con las predicciones de un modelo de regresiÃ³n lineal que se muestran en negro. ***El modelo lineal no logra capturar la tendencia no lineal exhibida por los datos***. A la derecha, tenemos el mismo diagrama de dispersiÃ³n junto con una lÃ­nea roja que corresponde a las predicciones del Ã¡rbol de regresiÃ³n que entrenÃ³ anteriormente. ***El Ã¡rbol de regresiÃ³n muestra una mayor flexibilidad y es capaz de capturar la no linealidad, aunque no completamente***. #### Entrenar el primer Ã¡rbol de regresiÃ³n
Entrenaremos un Ã¡rbol de regresiÃ³n para predecir el consumo mpg (millas por galÃ³n) de los automÃ³viles en el conjunto de datos de `mpg` automÃ¡tico utilizando las seis funciones disponibles.

El conjunto de datos se procesa por usted y se divide en un 80 % de entrenamiento y un 20 % de prueba.

```python
# Dividir variables dependientes de independiente
X = auto.drop(['mpg','origin'], axis = 1).values
y = auto['mpg'].values

# Dividir el conjunto en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.2,
                                                   random_state = 3)

# Instanciar dt
dt = DecisionTreeRegressor(max_depth=8,
             min_samples_leaf=0.13,
            random_state=3)

# Ajustar dt al conjunto de entrenamiento
dt.fit(X_train, y_train)
```

#### Evaluar el Ã¡rbol de regresiÃ³n
Evaluaremos el rendimiento del conjunto de pruebas `dt` utilizando la mÃ©trica del error cuadrÃ¡tico medio (RMSE). El RMSE de un modelo mide, en promedio, en quÃ© medida difieren las predicciones del modelo de las etiquetas reales. El RMSE de un modelo se puede obtener calculando la raÃ­z cuadrada del error cuadrÃ¡tico medio (MSE) del modelo.

```python
# Calcular y_pred
y_pred = dt.predict(X_test)

# Calcular mse_dt
mse_dt = MSE(y_test, y_pred)

# Calcular rmse_dt
rmse_dt = mse_dt**(1/2)

# Imprimir rmse_dt
print("RMSE del conjunto de prueba de dt: {:.2f}".format(rmse_dt))
```

#### RegresiÃ³n lineal vs Ã¡rbol de regresiÃ³n
Compararemos el RMSE del conjunto de pruebas `dt` con el logrado mediante un modelo de regresiÃ³n lineal. Ya hemos creado una instancia de un modelo de regresiÃ³n lineal `lr` y lo hemos entrenado en el mismo conjunto de datos que `dt`.

```python
# Instanciar modelo lineal
lr = LinearRegression()

# Ajustar el modelo a los datos
lr.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba
y_pred_lr = lr.predict(X_test)

# Calcular mse_lr
mse_lr = MSE(y_test, y_pred_lr)

# Calcular rmse_lr
rmse_lr = mse_lr**(1/2)

# Imprimir rmse_lr
print('RMSE del conjunto de prueba de la RegresiÃ³n Lineal: {:.2f}'.format(rmse_lr))

# Imprimir rmse_dt
print('RMSE del conjunto de prueba del Ãrbol de RegresiÃ³n: {:.2f}'.format(rmse_dt))
```
