### Tendencia
- La tendencia est presente cuando los puntos inicial y final son diferentes o tienen una pendiente ascendente o descendente en una serie de tiempo.

```python
plt.figure(figsize=(10, 6))
plt.plot(SP_prices)
plt.title('S&P-500 Prices')
plt.ylabel('$')
plt.xlabel('Date')
plt.show()
```

- Trazar una linea no es edecuado.
- Por lo que existe dos conceptos estadsticos:
    1. Funcin de autocorrelacin (ACF).- Analiza la relacin entre el valor actual de una serie de tiempo y sus valores rezagados.
        $$\hat{\rho} (h) = \frac{Cov(X_t,X_{t-h})}{Var(X_t)}$$
       Funciona mejor con MA

```python
# Dibujando la funcin de autocorrelacin
sm.graphics.tsa.plot_acf(SP_prices,lags=30)
plt.xlabel('Number of Lags')
plt.show()
```

    2. Funcin de autororrelacin parcial (PACF).- No deca rpidamente si no que se acerca a 0. Funciona mejor con AR(p)
        - Proporciona informacin sobre la correlacin entre el valor actual de una serie temporal y sus valores rezagados controlando las otras correlaciones.
        
        $$\hat{\rho}(h) = \frac{\text{Cov}(X_t, X_{t-h} | X_{t-1}, X_{t-2}, \ldots, X_{t-h-1})}{\sqrt{\text{Var}(X_t | X_{t-1}, \ldots, X_{t-h-1}) \text{Var}(X_{t-h} | X_{t-1}, \ldots, X_{t-h-1})}}$$

```python
# Trazar la funcin de autocorrelacin parcial
sm.graphics.tsa.plot_pacf(SP_prices, lags=30)
plt.xlabel('Number of Lags')
plt.show()
```

- Nos crentramos en los picos fuera de la regin de intervalo de confianza. Si por ejemplo el rezado 10 estuviera fuera, tendramos que tomar 10 rezagos.

#### Estacionariedad
Significa que las propiedades estadsticas como la media, la varianza y la covarianza de la serie temporal no cambian en el tiempo. Hay dos formas de estacionariedad:
1. Estacionariedad dbil: se dice que la serie temporal, $X_t$, es estacionaria si
    - $X_t$ tiene varianza finita, $E(X_t^2)<\infty, \forall t\in \mathbb{Z}.$
    - El valor medio de $X_t$, es constante y depende unicamnte del tiempo, $E(X_t) = \mu, \forall 
 t\in \mathbb{Z}$.
    - La estructura de covarianza, $\gamma(t,t+t)$, depende nicamente de la dfierencia de tiempo:
      $\gamma(h)=\gamma_h+\gamma(t+h,t)$.
      
En palabras, las series de tiempo deben tener una varianza finita con una media constante y una estructura de covarianza que sea funcin de la diferencia temporal.

2. Estacionariedad fuerte: Si la distribucin conjunta de $X_{t1}, X_{t2}, \cdots, X_{tk}$ es lo mismo con la versin desplazada del conjunto $X_{t1+h}, X_{t2+h}, ldots, X_{tk+h}$, se denomina estacionariedad fuerte. Por lo tanto, una fuerte estacionariedad implica que la distribucin de la variable aleatoria de un proceso aleatorio es la misma con un ndice de tiempo cambiante.

Debemos tomar encuenta la estacionariedad por:

1. En el proceso de estimacin es esencial tener cierta distribucin a medida que pasa el tiempo. En otras palabras, si la distribucin de una serie temporal cambia con el tiempo, se vuelve impredecible y no se puede modelar.

El objetivo final de los modelos de series temporales es la previsin. Para ello primero debemos estimar los coeficientes, lo que corresponde al aprendizaje en Machine Learning. Una vez que aprendemos y realizamos un anlisis de pronstico, asumimos que la distribucin de los datos en la estimacin permanece igual de manera que tenemos los mismos coeficientes estimados. Si este no es el caso, deberamos volver a estimar los coeficientes porque no podemos pronosticar con los coeficientes estimados anteriormente. 

***Tener un quiebre estructural como una crisis financiera genera un cambio en la distribucin. Debemos ocuparnos de este perodo con cautela y por separado**^. 

2. La otra razn para tener datos de estacionariedad es que, por supuesto, algunos modelos estadsticos requieren datos de estacionariedad, pero eso no significa que algunos modelos solo requieran datos estacionarios. En cambio, todos los modelos requieren estacionarios, pero si alimenta el modelo con datos no estacionarios, algunos modelos, por diseo, lo convierten en estacionario y lo procesan. Las funciones integradas en Python facilitan este anlisis de la siguiente manera:

```python
# Prueba de estacionariedad del ADF
stat_test = adfuller(SP_prices)[0:2]

# Impresin de la estadstica de prueba de los primeros cuatro decimales y el valor p de la prueba ADF
print("The test statistic and p-value of ADF test are {}".format(stat_test))
```

Existen, en general, dos formas de entender la no estacionariedad: visualizacin y mtodo estadstico. Este ltimo, por supuesto, tiene una forma mejor y ms slida de detectar la no estacionariedad. Sin embargo, para mejorar nuestra comprensin, comencemos con el ACF. ***El ACF de lenta decadencia implica que los datos no son estacionarios porque presentan una fuerte correlacin en el tiempo***. Eso es lo que observo en los datos del S&P-500. La forma estadstica de detectar la no estacionariedad es ms confiable y la prueba ADF es una prueba ampliamente apreciada. Segn el resultado de la prueba anterior, el valor p de 0,99 sugiere que los datos no son estacionarios, algo que debemos abordar antes de seguir adelante. 

Tomar diferencias es una tcnica eficaz para eliminar la no estacionariedad. Tomar diferencia no es ms que restar el valor actual de la serie de su primer valor rezagado***, es decir, $x_t  x_{t1}$ y el siguiente cdigo Python presenta cmo aplicar esta tcnica:

```python
# Tomando la diferencia de los precios del S&P-500
diff_SP_price = SP_prices.diff()

plt.figure(figsize=(10, 6))
plt.plot(diff_SP_price)
plt.title('Differenced S&P-500 Price')
plt.ylabel('$')
plt.xlabel('Date')
plt.show()

sm.graphics.tsa.plot_acf(diff_SP_price.dropna(),lags=30)
plt.xlabel('Number of Lags')
plt.show()

# Resultado de la prueba del ADF basado en datos diferenciados del S&P-500
stat_test2=adfuller(diff_SP_price.dropna())[0:2]
print("The test statistic and p-value of ADF test after differencing are {}".format(stat_test2))
```

- La tendencia no es el nico indicador de no estacionariedad.
- Para eliminar la estacionalidad, primero aplicamos el mtodo de remuestreo para calcular la media anual, que se utiliza como denominador en la siguente frmula:
  $$ndice\, estacional = \dfrac{Valor \,de\, una\, serie\, temporal\, estacional}{Promedio \,estacional}$$

```python
# Clculo de la media trimestral de utilizacin de energa. 
seasonal_index = energy.resample('Q').mean()

# Definicin de los aos en los que se ejecuta el anlisis de estacionalidad. 
dates = energy.index.year.unique()
deseasonalized = []
# Calcular el numerador de la frmula del ndice estacional. 
for i in dates:
    for j in range(1, 13):
        deseasonalized.append((energy[str(i)][energy[str(i)].index.month==j]))
        # Concatenando el uso de energa desestacionalizado. 
        concat_deseasonalized = np.concatenate(deseasonalized)
        
# Calcular el ndice estacional utilizando una frmula predefinida.
deseason_energy = []
for i,s in zip(range(0, len(energy), 3), range(len(seasonal_index))):
    # Se divide los datos mensuales por el ndice estacional trimestral para obtener la energa desestacionalizada.
    deseason_energy.append(concat_deseasonalized[i:i+3] / seasonal_index.iloc[s])
concat_deseason_energy = np.concatenate(deseason_energy)
deseason_energy = pd.DataFrame(concat_deseason_energy, index=energy.index)
deseason_energy.columns = ['Deaseasonalized Energy']
deseason_energy.head()

sm.graphics.tsa.plot_acf(deseason_energy, lags=10)
plt.xlabel('Number of Lags')
plt.show()

sm.graphics.tsa.plot_pacf(deseason_energy, lags=10)
plt.xlabel('Number of Lags')
plt.show()
```

## Modelos de series de tiempo
Los modelos tradicionales de series de tiempo son modelos univariados y siguen las siguientes fases: 
- Identificacin: En este proceso exploramos los datos utilizando ACF, PACF, identificando patrones y realizando pruebas estadsticas.
- -Estimacin: esta parte pertenece a la estimacin de coeficientes mediante una tcnica de optimizacin adecuada.
- -Diagnstico: Una vez estimada, debemos comprobar si los criterios de informacin o ACF/PACF sugieren que el modelo es vlido. Si es as, pasamos a la etapa de previsin.
- Pronstico: esta parte trata ms sobre el rendimiento del modelo. Al pronosticar, predecimos valores futuros basados en nuestra estimacin.

Mi objetivo, al modelar series de tiempo, es ajustar un modelo y comprender el carcter estadstico de una serie de tiempo, que flucta aleatoriamente en el tiempo. Recuerde la discusin sobre el proceso IID, que es el modelo de series de tiempo ms bsico y a veces se lo denomina ruido blanco.

### Ruido blanco
Se dice que la serie temporal $\epsilon_t$ es ruido blanco si satisface lo siguiente:
$$\epsilon_t\sim WN(0,\sigma^2_\epsilon).$$
$$Corr(\epsilon_t,\epsilon_s)=0, \; \forall t\neq s.$$
Es decir, $\epsilon_t$ tiene media 0 y varianza constante. Adems, no existe correlacin entre trminos sucesivos de $\epsilon_t$.

```python
mu = 0
std = 1
WN = np.random.normal(mu, std, 1000)
plt.plot(WN)
plt.xlabel('Number of Simulation')
plt.show()
```

A partir de este momento, ***necesitamos identificar el nmero ptimo de retraso antes de ejecutar el modelo de series de tiempo. Como puede imaginar, decidir el nmero ptimo de retraso es una tarea desafiante***. 

***Los mtodos ms utilizados son: ACF, PACF y criterios de informacin. Se han debatido el ACF y el PACF y, en aras de la exhaustividad, tambin se introducirn criterios de informacin, abreviado AIC.***

### Criterio de informacin (AIC)
Necesitamos tener un criterio para decidir qu modelo se ajusta mejor a los datos, ya que puede haber numerosos modelos potencialmente buenos. Los Criterios de Informacin de Akaike, tambin conocidos como AIC, como indican Cavanaugh y Neath (2019), AIC se introduce como una extensin del Principio de Mxima Verosimilitud. La mxima verosimilitud se aplica convencionalmente para estimar los parmetros de un modelo una vez que se han formulado la estructura y la dimensin del modelo. AIC se puede definir matemticamente como: 
$$AIC = 2ln(MaximumLikelihood) + 2d$$ 
donde $d$ es el nmero total de parmetros. El ltimo trmino, el $2d4 de la ecuacin, tiene como objetivo reducir el riesgo de sobreajuste. Tambin se denomina trmino de penalizacin mediante el cual puedo filtrar la redundancia innecesaria en el modelo. 

BIC es el otro criterio de informacin utilizado para seleccionar el mejor modelo. El plazo de sancin en BIC es mayor que el de AIC. 
$$BIC = 2ln(MaximumLikelihood) + ln(n) * d$$ 
donde $n4 es el nmero de observaciones.

### Modelo de medias mviles
La media mvil (MA) y los residuos son modelos estrechamente relacionados. La media mvil puede considerarse un modelo de suavizado, ya que tiende a tener en cuenta los valores de rezago del residual. En aras de la simplicidad, comencemos con 
$$MA(1): X_t = \epsilon_t + \alpha\epsilon_{t1}.$$ 
Siempre que $\alpha \neq 0$, tiene una estructura de correlacin no trivial. Intuitivamente, MA(1) nos dice que la serie temporal ha sido afectada por $\epsilon_t$ y $\epsilon_{t1}$ nicamente. En forma general, la ecuacin MA(q) se convierte en: 
$$X_t = \epsilon_t + \alpha_1\epsilon_{t1} + \alpha_2\epsilon_{t2}+\cdots + \alpha_q\epsilon_{tq}$$ 
A partir de este momento, para ser coherentes, modelaremos los datos de dos importantes empresas de TI, a saber Apple y Microsoft. Yahoo Finance proporciona una herramienta conveniente para acceder al precio de cierre de las acciones relacionadas para el perodo 01-01-2019 y 01-01-2021. Como primer paso, eliminamos los valores faltantes y comprobamos si los datos son estacionarios y resulta que ni los precios de las acciones de Apple ni de Microsoft tienen una estructura estacionaria como se esperaba. Por lo tanto, los pasos a seguir en este punto son tomar la primera diferencia para hacer que estos datos sean estacionarios y dividirlos como `trian` y `test`. El siguiente cdigo nos muestra cmo podemos hacer esto en Python.

```python
# Recuperar los precios de cierre mensuales de las acciones. 
ticker = ['AAPL', 'MSFT']
start = datetime.datetime(2019, 1, 1)
end = datetime.datetime(2021, 1, 1)
stock_prices = yf.download(ticker, start=start, end = end, interval='1d').Close

stock_prices = stock_prices.dropna()

for i in ticker:
    stat_test = adfuller(stock_prices[i])[0:2]
    print("The ADF test statistic and p-value of {} are {}".format(i,stat_test))

diff_stock_prices = stock_prices.diff().dropna()
# Dividir datos en 95% y 5%. 
split = int(len(diff_stock_prices['AAPL'].values) * 0.95)

# Asignar el 95% de los datos del precio de las acciones de Apple al conjunto de Train. 
diff_train_aapl = diff_stock_prices['AAPL'].iloc[:split]

# Asignar el 5% de los datos del precio de las acciones de Apple al conjunto de Test. 
diff_test_aapl = diff_stock_prices['AAPL'].iloc[split:]

# Asignar el 95% de los datos del precio de las acciones de Microsoft al conjunto de Train. 
diff_train_msft = diff_stock_prices['MSFT'].iloc[:split]

# Asignar el 5% de los datos del precio de las acciones de Microsoft al conjunto de Test. 
diff_test_msft = diff_stock_prices['MSFT'].iloc[split:]

# Guardar los datos para uso futuro.
# diff_train_aapl.to_csv('diff_train_aapl.csv')
# diff_test_aapl.to_csv('diff_test_aapl.csv')
# diff_train_msft.to_csv('diff_train_msft.csv')
# diff_test_msft.to_csv('diff_test_msft.csv')

fig, ax = plt.subplots(2, 1, figsize=(10, 6))
plt.tight_layout()
sm.graphics.tsa.plot_acf(diff_train_aapl,lags=30, ax=ax[0], title='ACF - Apple')
sm.graphics.tsa.plot_acf(diff_train_msft,lags=30, ax=ax[1], title='ACF - Microsoft')
plt.show()
```

- Al observar el primer panel de la Figura, se puede observar que hay picos significativos en algunos rezagos y, para estar seguros, elegimos el rezago 9 para el modelo de promedio mvil corto y 22 para el modelo de promedio mvil largo. Esto implica que el orden 9 ser nuestro orden a corto plazo y 22 se convertir en nuestro orden a largo plazo al modelar MA.

```python
# Promedio mvil con ventana corta para las acciones de Apple 
short_moving_average_appl = diff_train_aapl.rolling(window=9).mean()

# Promedio mvil con ventana larga para las acciones de Apple 
long_moving_average_appl = diff_train_aapl.rolling(window=22).mean()

# Grfico de lneas de los primeros precios diferenciados de las acciones de Apple 
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(diff_train_aapl.loc[start:end].index,
    diff_train_aapl.loc[start:end],
    label='Stock Price', c='y')
# Visualizacin del resultado de la media mvil de ventana corta para Apple 
ax.plot(short_moving_average_appl.loc[start:end].index,
    short_moving_average_appl.loc[start:end],
    label = 'Short MA',c='g')
# Visualizacin del resultado de la media mvil de ventana larga para Apple
ax.plot(long_moving_average_appl.loc[start:end].index,
    long_moving_average_appl.loc[start:end],
    label = 'Long MA',c='r')
ax.legend(loc='best')
ax.set_ylabel('Price in $')
ax.set_title('Stock Prediction-Apple')
plt.show()
```

## Modelo autoregresivo
Bsicamente pronosticamos el valor actual de la serie temporal $X_t$ utilizando una combinacin lineal de sus valores pasados. Matemticamente, la forma general de AR(p) se puede escribir como:
$$X_t = c+\alpha_1X_{t-1}+\alpha_2X_{t-2}+\cdots + \alpha_p X_{t-p}+\epsilon_t$$
donde $\epsilon_t$ denota el residual y $c$ el termino intercepto.

El modelo AR(p) implica que los valores pasados hasta el orden p tienen cierto poder explicativo sobre $X_t$. **Hemos discutido una de las propiedades principales de las series de tiempo, que es la estacionariedad, y la otra propiedad importante es la invertibilidad. Despus de presentar el modelo AR, es hora de mostrar la invertibilidad del proceso MA. Se dice que es invertible si se puede convertir a un modelo AR infinito**.

Bajo algunas circunstancias, MA puede escribirse como un proceso AR infinito, tal que deben tener una estructura de covarianza estacionaria, una parte determinista y un proceso MA invertible. Al hacer esto tenemos otro modelo llamado AR infinito gracias al supuesto de $|\alpha|<1$.

$$
\begin{array}{rcl}
X_t &=& \epsilon_t+\alpha\epsilon_{t-1}\\\\
    &=& \epsilon_t + \alpha\left(X_{t-1}-\alpha\epsilon_{t-2}\right)\\\\
    &=& \epsilon_t + \alpha X_{t-1}-\alpha^2\epsilon_{t-2}\\\\
    &=& \epsilon_t+\alpha X_{t-1}-\alpha^2\left(X_{t-2}+\alpha\epsilon_{t-3}\right)\\\\
    &=& \epsilon_t+\alpha X_{t-1}-\alpha^2 X_{t-2}+\alpha^3 \epsilon_{t-3}\\\\
    &=& \ldots\\\\
    &=& \alpha X_{t-1}-\alpha^2X_{t-2}+\alpha^3\epsilon_{t-3}-\alpha^4\epsilon_{t-3}+\cdots-(-\alpha)^n \epsilon_{t-n}
\end{array}
$$

De donde:
$$\alpha^n\epsilon_{t-n} = \epsilon_t-\sum_{i=0}^{n-1}\alpha^i X_{t-i}$$
Ahora, si $|\alpha|<1$. Entonces, $n\to \infty$
$$E\left(\epsilon_t - \sum_{i=0}^{n-1}\alpha^i X_{t-i}\right)^2 = E\left(\alpha^{2n}\epsilon_{t-n}^2 \to \infty\right)$$

Finalmente el proceso MA(1) resulta ser:
$$\epsilon_t = \sum_{i=0}^{\infty}\alpha^i X_{t-1}$$

Debido a la dualidad entre los procesos AR y MA, es posible representar AR(1) como MA infinita, MA($\infty$). En otras palabras, el proceso AR(1) puede expresarse como una funcin de los valores pasados de las innovaciones.
$$
\begin{array}{rcl}
    X_t &=& \epsilon_t+\theta X_{t-1}\\\\
        &=& \theta(\theta X_{t-2}+\epsilon_{t-1}+\epsilon_t\\\\
        &=& \theta^2 X_{t-2}+\theta \epsilon_{t-1} + \epsilon_t\\\\
        &=& \theta^2(\theta X_{t-3}+\theta \epsilon_{t-2}) \theta \epsilon_{t-1}+\epsilon_t\\\\
\end{array}
$$
$$X_t=\epsilon_t+\epsilon_{t-1}+\theta^2 \epsilon_{t-2}+\ldots + \theta^t X_t$$
Cmo $n\to \infty$, $\theta^t \to 0$, entonces puedeo representar AR(1) como un proceso MA infinito.

En el siguiente anlisis, ejecutamos un modelo autorregresivo para predecir los precios de las acciones de Apple y Microsoft. A diferencia de la media mvil, ***la funcin de autocorrelacin parcial es una herramienta til para encontrar el orden ptimo en un modelo autorregresivo***. Esto se debe a que, en AR, nuestro objetivo es descubrir la relacin de una serie de tiempo entre dos tiempos diferentes, digamos $X_t y $X_{t-k}$, y para hacerlo necesito filtrar el efecto de otros retrasos intermedios.

```python
sm.graphics.tsa.plot_pacf(diff_train_aapl,lags=30)
plt.title('PACF of Apple')
plt.xlabel('Number of Lags')
plt.show()

sm.graphics.tsa.plot_pacf(diff_train_msft,lags=30)
plt.title('PACF of Microsoft')
plt.xlabel('Number of Lags')
plt.show()
```

## Modelo de media mvil integrada autorregresiva
Es una funcin de los valores pasados de una serie temporal y ruido blanco. Incluso si incluimos datos no estacionarios, ARIMA los hace estacionarios al definir adecuadamente el parmetro. 
$$X_t = \alpha_1 d X_{t-1} + \alpha_2 d X_{t-1} + \cdots + \alpha_p d X_{t-p} + \epsilon_t + \beta_1 d \epsilon_{t-1} + \beta_2 d \epsilon_{t-2} + \cdots + \beta_q d \epsilon_{t-q}$$
donde $d$ hace refierencia a las diferencias.

### PROS
- ARIMA nos permite trabajar con datos sin procesar sin considerar si son estacionarios.
- Funciona bien con datos de alta frecuencia (a corto plazo).
- Es menos sensible a la fluctuacin de los datos en comparacin con otros modelos.

### CONTRAS
- ARIMA podra no captar la estacionalidad.
- Funciona mejor con series largas y datos a corto plazo (diarios, horarios).
- Como no se produce ninguna actualizacin automtica en ARIMA, no se debe observar ninguna interrupcin estructural durante el perodo de anlisis.
- No tener ajustes en el proceso ARIMA genera inestabilidad.

Ahora, mostraremos cmo funciona y se desempea ARIMA utilizando las mismas acciones, es decir, Apple y Microsoft.

```python
split = int(len(stock_prices['AAPL'].values) * 0.95)
train_aapl = stock_prices['AAPL'].iloc[:split]
test_aapl = stock_prices['AAPL'].iloc[split:]
train_msft = stock_prices['MSFT'].iloc[:split]
test_msft = stock_prices['MSFT'].iloc[split:]

# Configurando el modelo ARIMA para acciones de Apple. 
arima_aapl = ARIMA(train_aapl,order=(9, 1, 9))

# Adaptacin del modelo ARIMA al precio de las acciones de Apple. 
arima_fitted_aapl = arima_aapl.fit()

# Configurando el modelo ARIMA para acciones de Apple. 
arima_msft = ARIMA(train_msft, order=(6, 1, 6))

# Ajuste del modelo ARIMA al precio de las acciones de Microsoft. 
arima_fitted_msft = arima_msft.fit()

# Prediccin de los precios de las acciones de Apple segn ARIMA. 
arima_predictions_aapl = arima_fitted_aapl.predict(start=len(train_aapl),
end=len(train_aapl) + len(test_aapl) - 1, dynamic=False)

# Prediccin de los precios de las acciones de Microsoft segn ARIMA. 
arima_predictions_msft = arima_fitted_msft.predict(start=len(train_msft),
end=len(train_msft) + len(test_msft) - 1, dynamic=False)

# ndice de formacin para predicciones.
arima_predictions_aapl = pd.DataFrame(arima_predictions_aapl)
arima_predictions_aapl.index = diff_test_aapl.index
arima_predictions_msft = pd.DataFrame(arima_predictions_msft)
arima_predictions_msft.index = diff_test_msft.index


fig, ax = plt.subplots(2, 1, figsize=(18, 15))
ax[0].plot(diff_test_aapl, label='Actual Stock Price', c='b')
ax[0].plot(arima_predictions_aapl, c='r', label="Prediction")
ax[0].set_title('Predicted Stock Price-Apple')
ax[0].legend(loc='best')
ax[1].plot(diff_test_msft, label='Actual Stock Price', c='b')
ax[1].plot(arima_predictions_msft, c='r', label="Prediction")
ax[1].set_title('Predicted Stock Price-Microsoft')
ax[1].legend(loc='best')
for ax in ax.flat:
    ax.set(xlabel='Date', ylabel='$')
plt.show()
```

- Vale la pena discutir el mtodo alternativo para la seleccin de rezagos ptimos para modelos de series de tiempo. AIC es el mtodo que aplico aqu para seleccionar el nmero adecuado de retrasos. Tenga en cuenta que, aunque el resultado del AIC sugiere (4, 0, 4), el modelo no converge con este orden. Entonces, en su lugar se aplica (4, 1, 4).

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
import numpy as np
import datetime as dt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
import itertools


def load_raw_data(ticker, start_date, end_date):
    price = yf.download(ticker, start_date, end_date)['Adj Close']
    ret = np.log(price / price.shift(1))
    ret = pd.DataFrame(ret)
    ret.dropna(inplace=True)
    return ret


def stationary_test(ret):
    stat_test = adfuller(ret)[0:2]
    print("The ADF test statistic and p-value are {}".format(stat_test))


def train_test_split(ret):
    split = int(len(ret.values) * 0.95)

    train_data = pd.DataFrame()
    train_data = ret.iloc[:split]

    test_data = pd.DataFrame()
    test_data = ret.iloc[split:]

    return train_data, test_data


def arima_order_determine(train_data):
        p = q = range(0, 9)
        d = range(0, 6)
        pdq = list(itertools.product(p, d, q))
        arima_result = []
        for param_set in pdq:
            try:
                arima = ARIMA(train_data, order=param_set)
                arima_fitted = arima.fit()
                arima_result.append(arima_fitted.aic)
            except:
                continue
        if not arima_result:  # Check if the list is empty
            print("No ARIMA model was successfully fit.")
            return None, None
        else:
            print('**' * 40)
            lowest_aic = min(arima_result)
            best_params = pdq[arima_result.index(lowest_aic)]
            print(f"The lowest AIC score is {lowest_aic:.4f} and the corresponding parameters are {best_params}")
            return lowest_aic, best_params


def arima_model(train_data, test_data):
    arima = ARIMA(train_data, order=(3, 0, 7))
    arima_fit = arima.fit()
    arima_predict = arima_fit.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1, dynamic=False)
    arima_predict = pd.DataFrame(arima_predict)
    arima_predict.index = test_data.index

    fig, ax = plt.subplots(figsize=(18, 15))
    ax.plot(test_data, label='Actual Stock Price', linestyle='--')
    ax.plot(arima_predict, label='Prediction', linestyle='solid')
    ax.set_title('Predicted Stock Price')
    ax.legend(loc='best')
    plt.show()
    print(arima_fit.summary())

    return arima_fit, arima_predict


if __name__ == '__main__':
    ticker = ['AAPL']
    start_date = '2015-01-01'
    end_date = '2021-01-01'
    ret_ = load_raw_data(ticker, start_date, end_date)
    stationary_test(ret_)
    train_data_, test_data_ = train_test_split(ret_)
    lowest_aic, best = arima_order_determine(train_data_)
    arima_fit_, arima_predict = arima_model(train_data_, test_data_)
```
