## Estimación del riesgo crediticio
Además de la **probabilidad de impago**, el riesgo crediticio tiene tres características definitorias:
- Exposición. Esto se refiere a una parte que posiblemente pueda incumplir o sufrir un cambio adverso en su capacidad para cumplir.

- Probabilidad. La probabilidad de que esta parte incumpla sus obligaciones.

- Índice de recuperación. Cuánto se puede recuperar si se produce un incumplimiento.

### Calificación interna IRB
IRB estima el riesgo crediticio.

$$Pérdida \;esperada = EAD \cdot LGD \cdot PD$$

- PD = Probabilidad de incumplimiento.
- LGD = Pérdida esperada $[0,1]$.
- EAD = Exposición en caso de incumplimiento.

***La parte más importante es estimar es modelar la probabilidad de impago***.

## Clasificación de riesgos
La clasificación de riesgos no es más que agrupar a los prestatarios con una solvencia crediticia similar. Al dividir a los prestatarios en distintos grupos en función de su riesgo, la clasificación de riesgos nos permite hacer predicciones precisas.

La agrupación de riesgos se puede lograr mediante diferentes métodos estadísticos, pero aplicaremos una técnica de agrupación para terminar con grupos homogéneos utilizando ***K-medias***.

La distancia al **centroide** es la parte importante de ***K-medias***. Se puede medir mediante diferentes métodos:

- Euclidiano
  $$\sqrt{\sum_{i=1}^n (p_i-q_i)^2}$$
- Minkowski
  $$\left(\sum_{i= 1}^n |p_i-q_i|^p\right)^{1/p}$$
- Manhattan
  $$\sqrt{\sum_{i=1}^n|p_i-q_i|}$$

El objetivo de la agrupación es minimizar la distancia entre el centroide y las observaciones. que es otra forma de decir que estamos minimizando la suma del error cuadrado entre el centroide y las observaciones:
$$\sum_{i=1}^K \sum:{x\in C_i} (C_i-x)^2$$

- x = Observaciones.
- C_i = Centroido de i-enesimo grupo.

Luego, aplicamos el algoritmo de maximización de expectativs (EM), detrás de nuestro agrupamiento. $E$, se regiere a la asignación de observaciones al centroide más cercado y $M$ denota la finalización del proceso de generación de datos mediante la actualización de los parámetros. El algoritmo funciona de la siguiente manera:

1. Elija k puntos aleatorios para que sean centroides.

2. Según la métrica de distancia elegida, calcule las distancias entre las observaciones y los $n$ centroides. Según estas distancias, asigne cada observación al grupo más cercano.

3. Actualizar los centros de clúster según la asignación.

4. Repita el proceso desde el paso 2 hasta que el centroide no cambie.

Ahora, aplicamos la agrupación de riesgos utilizando la agrupación de K-medias. Para decidir el número óptimo de clusters, se emplearán diferentes técnicas. Primero, utilizamos el método del **codo**, que se basa en la inercia.

La **inercia** se calcula como la suma de las distancias al cuadrado de las observaciones hasta su centroide más cercano. En segundo lugar, se introduce la puntuación **Silhouette** como herramienta para decidir el número óptimo de conglomerados. Esta puntuación toma un valor entre 1 y -1. Un valor de 1 indica que una observación está cerca del centroide correcto y está correctamente clasificada. Sin embargo, -1 muestra que una observación no está correctamente agrupada. La fortaleza de la puntuación Silhouette se basa en tener en cuenta tanto la distancia intraconglomerado como la distancia interconglomerado. La fórmula para la puntuación Silhouette es la siguiente:
$$Silhouette = \dfrac{x-y}{\max(x,y)}$$

- x = distancia media entre conglomerados.
- y = Distancia media dentro de los conglomerados.

El tercer método es el de **Calinski-Harabasz (CH)**, conocido como criterio de la razón de varianza. La fórmula del método CH es la siguiente:
$$CH = \dfrac{SS_B}{SS_w}\cdot \dfrac{N-k}{k-1}$$

- $SS_B$ = Variación entre grupos.
- $SS_W$ = Varianza dentro del conglomerado.
- $N$ = Número de observaciones.
- $k$ = Número de conglomerados.

Con esta información, ***buscamos una puntuación CH alta, ya que cuanto mayor (menor) sea la varianza entre conglomerados (varianza dentro del conglomerado), mejor será para encontrar el número óptimo de conglomerados.***

El enfoque final es el **análisis de brechas**. Tibshirani et al. (2001) propusieron una idea única mediante la cual podemos encontrar el número óptimo de clústeres en función de la distribución de referencia. Siguiendo las notaciones similares de Tibshirani et al., sea $d_{ii}e$ una distancia euclidiana entre $x_{ij}$ y $x_{i^{ej}}$ y sea $C_r$ el i-ésimo cluster que denota el número de obsrevaciones en el cluster $r$:

$$\sum_j (x_{ij}-x_{i^{ej}})^2$$

La suma de distancias por pares para todas las observaciones en el grupo $r$:
$$D_r = \sum_{i,i^e\in C_r}d_{i,i^e}$$

La suma de cuadrados dentro del grupo, $W_k$ es:
$$W_k = \sum_{r=1}^k \dfrac{1}{2_n^r} D_r$$

- n = Tamaño de la muestra.

Y la expectativa de $W_k$ es:
$$W_k = \log(pn/12) - (2/p) \log(k) + c$$

- p = Dimensión.
- k = centroides.

La decisión se toma cuando obtenemos el valor de brecha más alta.

1. VEREMOS AHORA EL MÉTODO DE CODO

```python
# Aplicar la estandarización con fines de escalamiento
scaler = StandardScaler()
scaled_credit = scaler.fit_transform(numerical_credit)

distance = []
for k in range(1, 10):
    # Ejecución del algoritmo K-means
    kmeans = KMeans(n_clusters=k) 
    kmeans.fit(scaled_credit)
    # Calcular inercia y almacenar en una lista denominadadistance
    distance.append(kmeans.inertia_) 

plt.plot(range(1, 10), distance, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method')
plt.show()
```

- ***Para encontrar el número óptimo de conglomerados, observamos la pendiente de la curva y decidimos el punto de corte en el que la curva se vuelve más plana, es decir, la pendiente de la curva se vuelve más baja***. A medida que se vuelve más plano, la inercia, que nos dice a qué distancia se encuentran los puntos dentro de un grupo, disminuye, lo cual es bueno para el propósito de agrupar. Por otro lado, a medida que permitimos que la inercia disminuya, el número de conglomerados aumenta, lo que complica el análisis. Teniendo en cuenta esa compensación, el criterio de parada es el punto donde la curva se vuelve más plana.

2. PUNTACIÓN SILHOUETTE
Se presenta la puntación en el eje $x$ para los grupos 2 a 10.

```python
fig, ax = plt.subplots(4, 2, figsize=(25, 20))
for i in range(2, 10):
    km = KMeans(n_clusters=i)
    q, r = divmod(i, 2)
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick',
                                      ax=ax[q - 1][r])
    visualizer.fit(scaled_credit)
    ax[q - 1][r].set_title("For Cluster_"+str(i))
    ax[q - 1][r].set_xlabel("Silhouette Score")
```
